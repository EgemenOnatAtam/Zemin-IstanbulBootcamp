# Korelasyon HesabÄ±

```python
# take the most correlated features with SalePrice
corr = df.corr()
corr.sort_values(["SalePrice"], ascending = False, inplace = True)
print(corr.SalePrice)

corr5 = df.corr().nlargest(6, 'SalePrice')['SalePrice'].index
corr5
```

# Top 5 correlated columns with sales prices
- OverallQual: Rates the overall material and finish of the house
- GrLivArea: Above grade (ground) living area square feet
- GarageCars: Size of garage in car capacity
- GarageArea: Size of garage in square feet
- TotalBsmtSF: Total square feet of basement area

```python
# make a heatmap of the most correlated features with SalePrice
corr = df[corr5].corr()
fig = px.imshow(corr)
fig.show()
```

```python
# 3D scatter plot of the most correlated features with SalePrice with plotly 
fig = px.scatter_3d(df, x="OverallQual", y="GrLivArea", z="GarageCars", color="SalePrice", opacity=0.7)
fig.show()

```

```python
# Box plot (for example, SalePrice distribution by OverallQual)
sns.boxplot(x='OverallQual', y='SalePrice', data=df)
plt.show()
```

```python
# scatter plot for example, SalePrice distribution by grlivarea  with plotly
fig = px.scatter(df, x="GrLivArea", y="SalePrice", title="SalePrice distribution by grlivarea")
fig.show()
```

```python
# Sales prices and year built with plotly
fig = px.scatter(df, x="YearBuilt", y="SalePrice", title="Sales prices and year built")
fig.show()
```

```python
#3D  scatter plot of YearBuilt, YrSold, SalePrice with plotly
fig = px.scatter_3d(df, x="YearBuilt", y="YrSold", z="SalePrice", color="SalePrice", opacity=0.7)
fig.show()
```

```python
# (YrSold-YearBuilt) vs  sales price with plotly
import plotly.express as px

df['YrSold-YearBuilt'] = df['YrSold'] - df['YearBuilt']   # calculate difference
fig = px.scatter(df, x="YrSold-YearBuilt", y="SalePrice", title="(YrSold-YearBuilt) vs sales price")
fig.show()

```
