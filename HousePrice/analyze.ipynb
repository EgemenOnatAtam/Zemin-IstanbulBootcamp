# House Pricing Data Seti Raporu
Bu rapor, ev fiyatlarına ilişkin veri seti üzerinde yürütülen bir analizi özetlemektedir. Veri seti, ev özellikleri ve satış fiyatları gibi değişkenler içermektedir. Analiz, Python programlama dili kullanılarak gerçekleştirilmiştir.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import warnings
warnings.filterwarnings("ignore", category=FutureWarning, module="numpy")

```

# Veri Ön İncelmesi
Analize başlamadan önce, veri setindeki eksik değerlerin ve diğer önemli özelliklerin incelenmesi gerekiyordu. Eksik verileri görselleştirdik.



```python
df =  pd.read_csv("house_price.csv")
df.head()
```

# Satış Fiyatları Dağılımı
Satış fiyatları, veri setindeki en önemli değişkenlerden biridir. İlk olarak, satış fiyatlarının histogramını çıkardık ve dağılımı inceledik. Aşağıdaki grafikte, satış fiyatları dağılımı gösterilmektedir.
Dağılımın normal dağılma yakın olduğunu söyleyebiliriz. 


```python
# histogram of sales price with plotly
fig = px.histogram(df, x="SalePrice",  nbins=100, title="Histogram of Sales Price")
fig.show()
```

```python
# outliers  of sales price with plotly
fig = px.box(df, x="SalePrice", title="Boxplot of Sales Price")
fig.show()
```

# Korelasyon Analizi
Korelasyon analizi, veri setindeki değişkenler arasındaki ilişkileri incelememizi sağlar. Bu analizi yapmak için, SalePrice değişkeni ile diğer değişkenler arasındaki korelasyonu hesapladık. İlk 5 korelasyonlu değişken aşağıdaki gibidir:
- OverallQual: 0.790982
- GrLivArea: 0.708624
- GarageCars: 0.640409
- GarageArea: 0.623431
- TotalBsmtSF: 0.613581

Bu değişkenlerin korelasyonunu görselleştirmek için, aşağıdaki grafikleri çıkardık.


```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
# take the most correlated features with SalePrice
corr = df.corr()
corr.sort_values(["SalePrice"], ascending = False, inplace = True)
print(corr.SalePrice)

corr5 = df.corr().nlargest(6, 'SalePrice')['SalePrice'].index
```

```python
# make a heatmap of the most correlated features with SalePrice
corr = df[corr5].corr()
fig = px.imshow(corr)
fig.show()
```

```python
# 3D scatter plot of the most correlated features with SalePrice with plotly 
fig = px.scatter_3d(df, x="OverallQual", y="GrLivArea", z="GarageCars", color="SalePrice", opacity=0.7)
fig.show()

```

# Yaş
Yaş değişkeninin satış fiyatları ile düşük bir korelasyonu vardır. Aşağıdaki grafikte, yaş ve SalePrice değişkenleri arasındaki korelasyon görülmekted

```python
# Sales prices and year built with plotly
fig = px.scatter(df, x="YearBuilt", y="SalePrice", title="Sales prices and year built")
fig.show()
```

```python
#3D  scatter plot of YearBuilt, YrSold, SalePrice with plotly
fig = px.scatter_3d(df, x="YearBuilt", y="YrSold", z="SalePrice", color="SalePrice", opacity=0.7)
fig.show()
```

```python
# (YrSold-YearBuilt) vs  sales price with plotly
import plotly.express as px

df['YrSold-YearBuilt'] = df['YrSold'] - df['YearBuilt']   # calculate difference
fig = px.scatter(df, x="YrSold-YearBuilt", y="SalePrice", title="(YrSold-YearBuilt) vs sales price")
fig.show()

```

# Mahalle
Mahalle değişkeni, satış fiyatları ile ilgili önemli bir faktördür. Bazı mahallelerin diğerlerinden daha yüksek fiyatlarla satıldığı görülmektedir. Aşağıdaki grafikte, NoRidge, NridgHt ve StoneBr mahallelerinin diğerlerinden daha yüksek fiyatlarla satıldığı görülmektedir.

```python
# neighborhood and sales price with plotly
fig = px.box(df, x="Neighborhood", y="SalePrice", color =  "Neighborhood" ,  title="Neighborhood and sales price")
fig.show()
```

# Havalandırma
Havalandırma değişkeni, evlerin satış fiyatları ile ilgili önemli bir faktördür. Havalandırması olan evlerin havalandırması olmayan evlere göre daha yüksek fiyatlarla satıldığı görülmektedir. Aşağıdaki grafikte, havalandırması olan evlerin satış fiyatları, havalandırması olmayan evlere göre daha yüksektir.

```python
# CentralAir and sales price with plotly
fig = px.box(df, y="CentralAir", x="SalePrice", color =  "CentralAir" ,  title="CentralAir and sales price")
fig.show()
```
